### Ejemplo 1
Si $\vec{A}=10\vec{a_x}-4\vec{a_y}+6\vec{a_z}$ y $\vec{B}=2\vec{a_x}+\vec{a_y}$, halle:
1. El componente de $\vec{A}$ a lo largo de $\vec{a_y}$
2. La magnitud de $3\vec{A}-\vec{B}$
3. Un vector unitario a lo largo de $\vec{A}+2\vec{B}$

A continuaci√≥n se muestra la grafica de los vectores $\vec{A}$ y $\vec{B}$:

```python
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111, projection='3d')
#ax.quiver(X, Y, Z, U, V, W, length=1, normalize=False, color = 'red', alpha = .6,arrow_length_ratio = .18, pivot = 'tail',
#          linestyles = 'solid',linewidths = 3)

ax.grid()
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([0, 10])

vA = [0, 0, 0, 10, -4, 6]
vB = [0, 0, 0, 2, 1, 0]

vec = np.array([vA, 
                vB])

ax.text(10, -4, 6, '$(10, -4, 6)$')
ax.text(2, 1, 0, '$(2, 1, 0)$')
ax.text(0, 0, 0, '$(0, 0, 0)$')

X, Y, Z, U, V, W = zip(*vec)
ax.quiver(X, Y, Z, U, V, W, length=1, normalize=False, color = 'red', alpha = .6,arrow_length_ratio = .18, pivot = 'tail',
          linestyles = 'solid',linewidths = 3)

# A 

point1 = [10, -4, 6]
point2 = [10, -4, 0]
vec1 = np.array([point1, point2])
X, Y, Z = zip(*vec1)
ax.plot(X, Y, Z, lw = 2, color = 'b', alpha = 0.7)

point1 = [0, 0, 0]
point2 = [10, -4, 0]
vec1 = np.array([point1, point2])
X, Y, Z = zip(*vec1)
ax.plot(X, Y, Z, lw = 2, color = 'b', alpha = 0.7)

# B 

point1 = [2, 1, 0]
point2 = [2, 1, 0]
vec1 = np.array([point1, point2])
X, Y, Z = zip(*vec1)
ax.plot(X, Y, Z, lw = 2, color = 'b', alpha = 0.7)

point1 = [0, 0, 0]
point2 = [2, 1, 0]
vec1 = np.array([point1, point2])
X, Y, Z = zip(*vec1)
ax.plot(X, Y, Z, lw = 2, color = 'b', alpha = 0.7)

plt.show()
```

```python
A = np.array([10,-4,6])
B = np.array([2,1,0])
```

```python
# Punto 1
A_y = A[1]
print('Componente de A a lo largo de a_y es {0:d}.'.format(A_y))
```

```python
# Punto 2
C = 3*A - B
C_norm = np.linalg.norm(C)
print("|C| = es {0:.2f}".format(C_norm))
```

```python
# Punto 3
C = A + 2*B
C_norm = np.linalg.norm(C)
a_C = C/C_norm
print("a_C = ", a_C, sep = '')
```

### Ejemplo 4

Dados los vectores $\vec{A}=3\vec{a_x}+4\vec{a_y}+\vec{a_z}$ y $\vec{B}=2\vec{a_y}-5\vec{a_z}$ encuentre el angulo entre $\vec{A}$ y $\vec{B}$

```python
A = np.array([3,4,1])
B = np.array([0,2,-5])
AdotB = np.dot(A,B)
A_norm = np.linalg.norm(A)
B_norm = np.linalg.norm(B)
angle_rad = np.arccos(AdotB/(A_norm*B_norm))
angle_grad = np.degrees(angle_rad)
print('Angulo entre A y B = {0:.2f}'.format(angle_grad))
```

### Ejemplo 5

Tres candidades de campo estan dadas por:

$$\vec{P}=2\vec{a_x}-\vec{a_z}$$

$$\vec{Q}=2\vec{a_x}-\vec{a_y}+2\vec{a_z}$$

$$\vec{R}=2\vec{a_x}-3\vec{a_y}+\vec{a_z}$$

Determine:
1. $\left( \vec{P}+\vec{Q} \right) \times \left( \vec{P}-\vec{Q} \right)$
2. $\vec{Q}\cdot\vec{R}\times\vec{P}$
3. $\vec{P}\cdot\vec{Q}\times\vec{R}$
4. $\sin{\left(\theta_{QR}\right)}$
5. $\vec{Q}\cdot\vec{R}\times\vec{P}$
6. $\vec{P} \times \left( \vec{Q} \times \vec{R} \right)$
7. Un vector unitario perpendicular tanto a $\vec{P}$ como a $\vec{R}$
8. El componente $\vec{P}$ a lo largo de $\vec{Q}$

```python
P = np.array([2,0,-1])
Q = np.array([2,-1,2])
R = np.array([2,-3,1])
```

```python
# Punto 1
P1 = P + Q
P2 = P - Q
Resp = np.cross(P1, P2)
print("Resp = ", Resp, sep = '')
```

```python
# Punto 2
Resp = np.dot(Q,np.cross(R,P))
print("Resp = ", Resp, sep = '')
```

```python
# Punto 3
Resp = np.dot(P,np.cross(Q,R))
print("Resp = ", Resp, sep = '')
```

```python
# Punto 4
V = np.cross(Q, R)
norm_V = np.linalg.norm(V)
norm_Q = np.linalg.norm(Q)
norm_R = np.linalg.norm(R)
resp = norm_V/(norm_Q*norm_R)
print("sen(theta_PQ) = {0:.4f}".format(resp))
```
